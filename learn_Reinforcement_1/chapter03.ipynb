# 梯度下降法


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
def f(x):
    return x * x - 2 * x + 1

def g(x):
    return 2 * x - 2
```

```python
x = np.linspace(-5,7,100)
y = f(x)

figsize(9, 5)
plt.plot(x, y, label= 'f(x)');
plt.legend();
plt.title('f(x)=(x-1)^2');
plt.xlabel('x');
plt.ylabel('f(x)');
```

```python
def gd(x_start, step, g): 
    """
    x_start : x的初始值
    step : 步长
    g : 求梯度的函数
    """
    li_x = []  # 记录每一步的x
    x = x_start
    
    for i in range(30):
        li_x.append(x)
        grad = g(x)
        x -= grad * step
        print( '[ Epoch {0} ] grad = {1}, x = {2}'.format(i, grad, x) )
        if abs(grad) < 1e-6:
            break;
    return li_x
```

```python
li_x = gd(5, 0.1, g)

plt.plot(li_x, label= 'x',marker='o');
plt.legend();
plt.title('step = 0.1');
plt.xlabel('Num of Iteration');
plt.ylabel('x');
```

```python
li_x = gd(5, 100, g)

plt.plot(li_x, label= 'x',marker='o');
plt.legend();
plt.title('step = 100');
plt.xlabel('Num of Iteration');
plt.ylabel('x');
```

```python
li_x = gd(5, 1, g)

plt.plot(li_x, label= 'x',marker='o');
plt.legend();
plt.title('step = 1');
plt.xlabel('Num of Iteration');
plt.ylabel('x');
```

```python
li_x = gd(4, 1, g)

plt.plot(li_x, label= 'x',marker='o');
plt.legend();
plt.title('step = 1');
plt.xlabel('Num of Iteration');
plt.ylabel('x');
```

# 动量算法

## 重要函数

```python
def gd(x_start, step, g):
    x = x_start
    xarr = [x]
    for i in range(20):
        grad = g(x)
        x -= grad * step
        xarr.append(x.copy())
        print ('[ Epoch {0} ] grad = {1}, x = {2}'.format(i, grad, x))
        if sum(abs(grad)) < 1e-6:
            break;
    return x, xarr
```

```python
def momentum(x_start, step, g, discount = 0.7):   
    x = np.array(x_start, dtype='float64')
    pre_grad = np.zeros_like(x)
    xarr = [x.copy()]
    for i in range(50):
        grad = g(x)
        pre_grad = pre_grad * discount + grad * step
        x -= pre_grad
        xarr.append(x.copy())
      
        print('[ Epoch {0} ] grad = {1}, x = {2}'.format(i, grad, x))
        if sum(abs(grad)) < 1e-6:
            break;
    return x, xarr
```

## Nesterov 算法 

```python
# 前面定义了，这里是用来对比
def momentum(x_start, step, g, discount = 0.7):   
    x = np.array(x_start, dtype='float64')
    pre_grad = np.zeros_like(x)
    xarr = [x.copy()]
    for i in range(50):
        grad = g(x)
        pre_grad = pre_grad * discount + grad * step
        x -= pre_grad
        xarr.append(x.copy())
      
        print('[ Epoch {0} ] grad = {1}, x = {2}'.format(i, grad, x))
        if sum(abs(grad)) < 1e-6:
            break;
    return x, xarr
```

```python
def nesterov(x_start, step, g, discount = 0.7):   
    x = np.array(x_start, dtype='float64')
    pre_grad = np.zeros_like(x)
    x_arr = [x.copy()]
    for i in range(50):
        x_future = x - step * discount * pre_grad
        grad = g(x_future)  # 不同处在于计算梯度的点x
        pre_grad = pre_grad * 0.7 + grad 
        x -= pre_grad * step
        x_arr.append(x.copy())
        
        print('[ Epoch {0} ] grad = {1}, x = {2}'.format(i, grad, x))
        if abs(sum(grad)) < 1e-6:
            break;
    return x, x_arr
```

# 共轭梯度法

我们将待优化的问题形式定义为
$$min_X \frac{1}{2}  X^TAX - b^TX$$

其中 $X$ 为待优化的向量 ， $A$ 为半正定的矩阵， $b$ 为已知向量 ，对公式进行求导，并令导数为 0 ，就可以得到
$$AX^* = b$$

其中 $X^*$为最优值。 从公式的形式来看 ， 虽然可以通过高斯消元法直接解方程，但计算量比较大 ，计算复杂度比较高 ，所以我们希望通过优化的方式找到最优解 。

误差：

$$e_t = x^* -x_t$$

现在令 $r_t$ 表示第 $t$ 轮的更新方向，就可以得到下面的公式:

$$ r^T_t e_{t+1} = 0 $$

假设每一轮的优化量都和误差正交 ，那么如果我们的优化空间有 $d$ 维，理论上最多只需要迭代 $d$ 轮就可以求解出来，这样在优化时间上就有了保证。 

如果我们直接使用这个公式，就会发现一个问题：公式中需要知道误差 。换句话说，我们需要知道最优点在哪儿。 如果我们知道最优点在哪儿，就不用优化了 。
可是不知道最优点儿，这个方法又无法直接使用，于是我们仿佛掉入了“先有鸡还是先有蛋”的陷阱 。


不过庆幸的是，我们还有数学武器在手，可以用线性代数的知识偷天换日，解决这个问题，让这个美好的约束发挥它的作用 。

## 共轭

前面我们提到优化方向和误差正交，如果使用了共辄这个工具，现在两者的关系将变为共轭正交，也就是存在一个矩阵 $A$($A$就是轭 ），使得优化方向和误差这两头牛“牵手成功”，满足正交的性质：

$$ r^T_t A e_{t+1} = 0 $$

看到这里可能有的读者依然很迷惑，为什么要加人这个矩阵，它的加入不就使原本正交的两项变得不正交了吗？别着急，其实如果我们在上面的原始公式中间加一个特殊的矩阵 ， 例如单位阵，看上去就舒服很多了：


$$ r^T_t I e_{t+1} = 0 $$



## 优化步长的确定

共辄梯度法属于线搜索的一种，因此和梯度下降法类似，我们的总体思路不变，优化过程分如下两步：
1. 确定优化方向 。
2. 确定优化步长 。

## 共轭梯度

P65页推导公式有问题


```python
# 代码推导见 P68
def cg(f_Ax, b, cg_iters=10, callback=None, verbose=False, residual_tol=1e-10):
    p = b.copy()  # 真正的优化方向p
    r = b.copy()  # 残差r，初始残差r1为b
    x = np.zeros_like(b)  # 初始值为0
    
    rdotr = r.dot(r)  # rT*r
    for i in range(cg_iters):
        z = f_Ax(p)
        v = rdotr / p.dot(z)  # 步长
        x += v*p
        r -= v*z
        newrdotr = r.dot(r)
        mu = newrdotr/rdotr
        p = r + mu*p
        rdotr = newrdotr
        
        if rdotr < residual_tol:
            break
    
    return x
```

# 自然梯度法

## Fisher 信息短阵

---
KL 散度是概率论和信息论中十分重要的一个概念 。 它是描述两个概率分布 P和Q 差异的一种方法

$$KL(p||q) = \sum p(x) log \frac{p(x)}{q(x)}$$
$$KL(p||q) = \int p(x) log \frac{p(x)}{q(x)}$$

KL 散度可以很好地测量两个概率分布之间的距离 。 两个分布越接近， KL 散度越小； 两个分布越远， KL 散度就越大 。

---

这里分别使用 KL 散度和 Fisher 信息矩阵近似两种方式计算另一个分布之间的距离 。 代码如下所示：

```python
from math import log
import matplotlib.pyplot as plt
import numpy as np

def kl(p,q):
    return p * log(p / q) + (1 - p) * log((1 - p) / (1 - q))

def fisher(p):
    return 1 / p / (1 - p)

def fisher_kl(p,q):
    return 0.5 * fisher(p) * (p - q)*(p - q)

x = np.linspace(0.1, 0.9, 20)
y1 = [kl(0.5, q) for q in x]
y2 = [fisher_kl(0.5, q) for q in x]

print('KL 散度与 Fisher 信息近似的对比图（ theta = 0.5 的伯努利分布与另一个伯努利分布的比较）')

plt.plot(x, y1, label='KL')
plt.plot(x, y2, label='Fisher_KL')
plt.ylabel('value')
plt.xlabel('theta_2')
plt.legend();
plt.show();
```

由于我们选取的参照点为 $ \theta = 0.5$ 的分布位置，可以看出在 0.5 附近， KL 散度和 Fisher 信息、矩阵近似的 KL 散度十分接近 。 因此我们确信在局部范围内两者是可以相互替换的 。 这就是 Fisher信息矩阵的实际意义 。

---

熟悉优化算法的读者在看到这个公式形式后，一定会联想到牛顿法 。 牛顿法是一个二阶梯度算法，它求解优化方向的公式为

$$ g = - \nabla^2_wf(w)^{-1} \nabla_wf(w) $$

我们很容易看出两个公式的区别与联系 。 

根据 Fisher 信息矩阵的求解方法，自然梯度法可以变成一个一阶优化问题，也可以变成一个二阶优化问题。 
- 如果把它看作一个一阶优化问题，那么需要对优化步长做更多的考量；
- 如果把它看作一个二阶优化问题，那么牛顿法中可能遇到的一些问题同样会在自然梯度法中出现。

实际上，自然梯度法是一个十分复杂的方法，本节只介绍基本原理，在第 10 章中，我们将继续介绍基于此方法进行优化的细节 。


# 总结

本节我们主要介绍了4种优化算法，让我们在此做一个总结。
1. 梯度下降法 ： 算法简单直接 ，但是在设计优化步长时需要认真考量 。
2. 动量法 ：更好地穿越一些平坦的优化区域。
3. 共辄梯度法 ： 强调每一步优化迭代的质量 。
4. 自然梯度法 ： 将每一轮迭代中对参数的更新转变为对模型效果的更新。
